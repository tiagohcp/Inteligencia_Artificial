```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Leitura dos dados do dataset
nba =  pd.read_csv('players_stats_new.csv')
nba.head()
```

```python
#Removendo as variÃ¡veis que nÃ£o tem interesse
nba = nba.drop(['MIN','FGA','FG%','3PM','3PA','FTM','FTA','OREB','DREB','BLK','TOV','EFF','AST/TOV','STL/TOV','Age',
                'Birth_Place','Birthdate','Collage','Experience','Height','Pos','Team','Weight','BMI'], axis=1)
#VariÃ¡veis de interesse 'Name','Games Played', 'PTS','FGM', '3P%', 'FT%', 'REB','AST','STL','PF'
nba
```

```python
# coeficiente de relaÃ§Ã£o (quanto mais prÃ³ximo de 1.0 ou -1.0, melhor atributo)
nba.corr()
```

```python
#SeleÃ§Ã£o dos dados numericos do dataset contidos nas colunas 1 Ã  9
Xnba = nba.iloc[:, 1:10].values
Xnba
```

```python
# Parametrizando K-Means com 3 Clusters e Metodo de InicializaÃ§Ã£o RandÃ´mico
kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
# Aplicando o K-Means
kmeans.fit(Xnba)
```

```python
# Listando os Centroids para cada um dos 04 Clusters
kmeans.cluster_centers_
```

```python
# Realcula os Centroids..
kmeans.fit(Xnba)
```

```python
# Listando os Centroids para cada um dos 04 Clusters
kmeans.cluster_centers_
```

```python
# Executa K-Means para agrupar os dados e retorna Tabela de Distancias
# Calcula a distancia entre os dados de cada linha (instÃ¢ncia) e cada um dos 4 clusters
distance = kmeans.fit_transform(Xnba)
distance
```

```python
# Para cada uma das instÃ¢ncias, a qual dos 4 clusters ela pertence..0, 1, 2 ou 3 ?
labels = kmeans.labels_
labels
```

```python
#Adiciona uma coluna 'cluster' ao dataframe para alocar a qual cluster a linha foi clasificada
nba['cluster'] = labels
nba
```

```python
# Centroides Calculados para cada uma das Features, para cada um dos 4 clusters
kmeans.cluster_centers_
```

```python
# Visualizando os Clusters
x = nba['Games Played']
y = nba['PTS']
plt.scatter(x, y, s = 50, c = nba['cluster'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 150, c = 'red')
plt.title('Players Clusters and Centroids')
plt.xlabel('Games Played')
plt.ylabel('Points')
plt.legend()
plt.show()
```

```python
# Visualizando os Clusters
x = nba['PTS']
y = nba['FGM']
plt.scatter(x, y, s = 50, c = nba['cluster'])
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s = 150, c = 'red')
plt.title('Players Clusters and Centroids')
plt.xlabel('Points')
plt.ylabel('Field Goals Made')
plt.legend()
plt.show()
```

```python
# Vamos agrupar uma nova instÃ¢ncia ?
# Dado um perfil de cliente, cliente
# O resultado indica a qual cluster esse novo cliente foi atribuÃ­do!
newcustomer = [ [70,750,280,30.8,75.9,380,150,66,146] ]
kmeans.predict(newcustomer)
```

```python
# Metododo de Elbow: como encontrar valor ideal de K ?
# Basicamente o que o mÃ©todo faz Ã© testar a variÃ¢ncia dos dados em relaÃ§Ã£o ao nÃºmero de cluters
# O que percebemos no GrÃ¡fico abaixo?
# Soma dos erros quadrÃ¡ticos de cada cluster (WCSS) cai e se estabiliza, Ã  medida que aumenta a qtd de clusters
wcss = []
maxclusters = 11
for i in range(1, maxclusters):
 kmeans = KMeans(n_clusters = i, init = 'random')
 kmeans.fit(Xnba)
 print(i,kmeans.inertia_)
 wcss.append(kmeans.inertia_)
plt.plot(range(1, maxclusters), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Qtde de Clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```
