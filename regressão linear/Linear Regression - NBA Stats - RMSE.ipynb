```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Medição do erro pelo método RMSE
#Aonde P são os dados da predição e X são os dados de verificação, n é quantidade de amostras.
def RMSE(pred, x):
    n = pred.shape[0]
    return np.sqrt(np.sum((pred-x)**2)/n)
```

```python
#Leitura da base de dados
nba = pd.read_csv('players_stats_new.csv')
nba
```

```python
#Seleciona o set de treinamento com as quantidadedes de arremessos e as quantidades de arremeços acertados
train = nba.iloc[0::2, 4:6].values
train
```

```python
#Seleciona o set de validação com as alturas e pesos
val = nba.iloc[1::2, 4:6].values
val
```

```python
# Plota tentativa X sucesso para os sets de treinamento e validação
%matplotlib inline

# Set de treinamento
plt.subplot(1,2,1)
plt.plot(train[:,0], train[:,1], linestyle='None', marker='o', markersize=5)
plt.title('Training')

# Set de validação
plt.subplot(1,2,2)
plt.plot(val[:,0], val[:,1], linestyle='None', marker='o', markersize=5)
plt.title('Validation')

plt.show()
```

```python
#Medida da correlação usando a Correlação Normalizada
def correlation(x, y):    
    x1 = (x - np.mean(x))
    y1 = (y - np.mean(y))
    
    num = np.sum(x1 * y1)
    div = np.sqrt( np.sum(x1 ** 2) * np.sum(y1 ** 2) )
    
    return num/div
```

```python
#Cálculo dos Coeficientes b0 e b1
b1 = correlation(train[:,0], train[:,1]) * np.std(train[:,1]) / np.std(train[:,0])
b0 = np.mean(train[:,1]) - (b1 * np.mean(train[:,0]))
print(b0, b1)
```

```python
#Calculo da predição do peso através da altura
def compute_single_made(attempt):
    return b0 + b1*attempt

print(compute_single_made(201))
```

```python
#Validação
pred_made = b0 + (b1 * val[:,0])
rmse = RMSE(pred_made, val[:,1])

print(rmse)

# Plot os pesos preditos de acordo com o set de validação
plt.plot(val[:,0], val[:,1], linestyle='None', marker='o', markersize=5)
plt.plot(val[:,0], pred_made)
plt.title('Linear Regression - Field Goals Made x Attempted')
plt.xlabel('Field Goals Attempted')
plt.ylabel('Field Goals Made')
```
